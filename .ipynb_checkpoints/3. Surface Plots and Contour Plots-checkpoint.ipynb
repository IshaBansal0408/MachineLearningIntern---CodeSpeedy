```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
plt.style.use('seaborn-poster')
plt.rcParams['font.family'] = "serif"
```

# Surface Plots

Data preparation

```python
# x and y data
x=np.arange(1,100,0.02)
y=x
a,b=np.meshgrid(x,y)
```

3-D Graph Structure

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
plt.title("Empty 3-D Plot Structure")
plt.show()
```

Plot - 1 

```python
z1=a+b
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(a,b,z1,cmap='gnuplot')
plt.title("z=x+y")
plt.show()
```

Plot - 2 

```python
a=np.arange(-1,1,0.02)
b=a
a,b=np.meshgrid(a,b)
z2=a**2+b**2
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(a,b,z2,cmap='gnuplot') #cmap - color map
plt.show()
```

Plot - 3

```python
z3=a**3+b**3
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(a,b,z3,cmap='gnuplot')
plt.title("z=x^3+y^3")
plt.show()
```

Plot - 4

```python
a1=np.linspace(-1,1)
b1=np.linspace(0,2*np.pi)
[x,y]=np.meshgrid(a1,b1)
z4=x*np.cos(y)
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(x,y,z4,cmap='gnuplot')
plt.title("z=x*cos(y)")
plt.show()
```

Plot - 5

```python
a2=np.linspace(-5,5,81)
b2=np.linspace(0,10,81)
[x,y]=np.meshgrid(a2,b2)
z5=3*np.sin(x)*np.cos(y)
```

```python
z5=3*np.sin(x)*np.cos(y)
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(x,y,z5,cmap='gnuplot')
plt.title("z=3*sin(x)*cos(y)")
plt.show()
```

# Contour Plots

```python
a=np.arange(-1,1,0.02)
b=a
a,b=np.meshgrid(a,b)
z1=a+b
z2=a**2+b**2
z3=a**3+b**3

plt.figure(figsize=(30,10))

plt.subplot(1,3,1)
plt.contour(a,b,z1,linestyles='dashed',linewidths=1,cmap='prism')
plt.title("z=x+y")
plt.xlabel("x values")
plt.ylabel("y values")

plt.subplot(1,3,2)
plt.contour(a,b,z2,linestyles='dashed',linewidths=1,cmap='prism')
plt.title("z=x^2+y^2")
plt.xlabel("x values")
plt.ylabel("y values")

plt.subplot(1,3,3)
plt.contour(a,b,z3,linestyles='dashed',linewidths=1,cmap='prism')
plt.title("z=x^3+y^3")
plt.xlabel("x values")
plt.ylabel("y values")

plt.show()
```

```python
a=np.arange(-1,1,0.02)
b=a
a,b=np.meshgrid(a,b)
z1=a+b
z2=a**2+b**2
z3=a**3+b**3

plt.figure(figsize=(30,10))

plt.subplot(1,3,1)
plt.contourf(a,b,z1,cmap='gnuplot')
plt.title("z=x+y")
plt.xlabel("x values")
plt.ylabel("y values")

plt.subplot(1,3,2)
plt.contourf(a,b,z2,cmap='gnuplot')
plt.title("z=x^2+y^2")
plt.xlabel("x values")
plt.ylabel("y values")

plt.subplot(1,3,3)
plt.contourf(a,b,z3,cmap='gnuplot')
plt.title("z=x^3+y^3")
plt.xlabel("x values")
plt.ylabel("y values")

plt.show()
```

```python
fig=plt.figure()
axes = fig.gca(projection='3d')
axes.contour(a,b,z1,cmap='prism')
fig=plt.figure()
axes = fig.gca(projection='3d')
axes.contour(a,b,z2,cmap='prism')
plt.title("z=x^2+y^2")
fig=plt.figure()
axes = fig.gca(projection='3d')
axes.contour(a,b,z3,cmap='plasma')
plt.title("z=x^3+y^3")
plt.show()
```
